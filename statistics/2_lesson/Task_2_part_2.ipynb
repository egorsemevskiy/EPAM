In this test we will check is there any relationship between Sex and PreferCar

## Define Hypothesis

Null Hypothesis (H0): Two variables are independent.

Alternate Hypothesis (H1): Two variables are not independent.

## chi^2 test

```python
crosstab, test_results, expected  = rp.crosstab(df_chi["Sex"], df_chi["PreferCar"],
                                                test= "chi-square",
                                                expected_freqs= True,
                                                prop= "cell")

crosstab
```
